```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sinasc = pd.read_csv('sinasc_RO_2019.csv')
sinasc
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
#forma mais elaborada

plt.rc('figure', figsize=(12,6))

fig, axes = plt.subplots(1,2)


# Utilizando a biblioteca seaborn:
sns.boxplot(y='IDADEMAE', data=sinasc, ax=axes[1])
axes[1].set_title(label="'IDADEMAE'")
axes[1].set_ylabel(ylabel='Idade da mãe (em anos)')

# Utilizando a biblioteca pandas:
sinasc.boxplot(column='IDADEMAE', ax=axes[0])
axes[0].set_title(label="'IDADEMAE'")
axes[0].set_ylabel(ylabel='Idade da mãe (em anos)')
axes[0].set_xticks(ticks=[1], labels='')
plt.subplots_adjust(wspace=.3)
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.close()
plt.rc('figure', figsize= (12,6))

fig, axes = plt.subplots(1,2)

# Utilizando a biblioteca seaborn:
sns.boxplot(y='PESO', data=sinasc, ax=axes[1])
axes[1].set_title(label="'PESO BEBÊ'")
axes[1].set_ylabel(ylabel='Peso do bebê (em gramas)')

# Utilizando a biblioteca pandas:
sinasc.boxplot(column='PESO', ax=axes[0])
axes[0].set_title(label="'PESO BEBÊ'")
axes[0].set_ylabel(ylabel='Peso do bebê (em gramas)')
axes[0].set_xticks(ticks=[1], labels='')
plt.subplots_adjust(wspace=.4)
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
# Utilizando a biblioteca pandas:

plt.close()
plt.rc('figure', figsize=(12,5))

fig, axes = plt.subplots(1,2)


# Utilizando a biblioteca pandas:
sinasc['IDADEMAE'].plot.hist(ax=axes[0])
axes[0].set_title(label="'IDADEMAE'")
axes[0].set_xlabel(xlabel='Idade da mãe (em anos)')

# Utilizando a biblioteca seaborn:
sns.histplot(data=sinasc['IDADEMAE'], bins=10, ax=axes[1], kde= True, alpha= .25)
axes[1].set_title(label="'IDADEMAE'")
axes[1].set_xlabel(xlabel='Idade da mãe (em anos)')

plt.subplots_adjust(wspace=.3)
plt.show()

#outra Forma mais simples seria
#sinasc['IDADEMAE'].plot.hist(bins = 20)
```
