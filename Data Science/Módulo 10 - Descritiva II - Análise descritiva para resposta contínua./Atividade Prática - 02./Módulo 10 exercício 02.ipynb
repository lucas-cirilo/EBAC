Considere a mesma base de dados utilizada na tarefa 1 ```previsao_de_renda.csv```.

1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
````

````python
df = pd.read_csv('previsao_de_renda.csv')
df = df.drop(columns='Unnamed: 0').set_index(keys='index', drop=True)
df.drop_duplicates(inplace=True)
df
````

````python
plt.rc('figure', figsize=(14,12))
fig, axes = plt.subplots(2,2)

sns.pointplot(x='posse_de_imovel',
              y='renda',
              data=df,
              dodge=True,
              ax=axes[0,0])

sns.pointplot(x='posse_de_imovel',
              y='renda',
              data=df,
              dodge=True,
              ax=axes[0,1])

sns.pointplot(x='posse_de_imovel',
              y='renda',
              data=df,
              hue = 'posse_de_veiculo',
              dodge=True,
              ax=axes[1,0])

sns.pointplot(x='posse_de_veiculo',
              y='renda',
              data=df,
              hue = 'posse_de_imovel',
              dodge=True,
              ax=axes[1,1])

````

**Alguma delas parece discriminar renda de forma estatisticamente significante?**

Sim, nota-se, na primeira linha, que ambas as variáveis qualitativas discriminam a renda de forma estatisticamente significante. Porém, a variável, **posse de Veículo** possui maior relevância na predição de renda.

Enquanto que os dados de **posse de imovel** possui menor relevância.
